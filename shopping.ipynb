```python
# Importing the appropriate libraries
import sklearn # Modelling
import pandas as pd # tables
import datetime
from sklearn import preprocessing
from sklearn.model_selection import train_test_split # Splitting testing training
from sklearn.neural_network import MLPClassifier
```

```python
# Read in the raw data
shoppingTable = pd.read_csv("juneShopping.csv")
```

```python
# Data pre-processing

# Keep only the date and the description
shoppingTable['created'] = pd.to_datetime(shoppingTable['created'])
shoppingTable['day_of_week'] = shoppingTable['created'].dt.weekday
X = shoppingTable['day_of_week']
Y = shoppingTable.filter(items=['description'])

YUnique = Y['description'].unique()

#le = preprocessing.LabelEncoder()
Y["description"] = Y["description"].astype('category')

# Decode table

Y["catcodes"] = Y["description"].cat.codes

Y
#le.fit_transform(Y['description'].values)
#le.fit(YUnique.values)
#le.transform(["Sainsbury's"])
#days = {0:'Waitrose',1:'Tesco',2:'Sainsbury\'s',3:'Friends House',4:'Poundland',5:'McDonald\'s',6:'Tesco'}
```

```python
# Partition the dataset
X_train, X_test, y_train, y_test = \
        train_test_split(X, Y["catcodes"], test_size=.4, random_state=42)
```

```python
clf = MLPClassifier();
clf.fit(X_train.to_frame(),y_train.to_frame())
score = clf.score(X_test.to_frame(), y_test.to_frame())

score
clf.predict(5)
```
